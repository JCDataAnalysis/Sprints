#### Descripció del Dataset 

El dataset conté 1.936.758 files i 30 columnes amb les variables següents:

- Year: 2008 en tots els casos
- Month: d'1 a 12
- DayofMonth: d'1 a 31
- DayOfWeek: d'1 (dilluns) a 7 (diumenge)
- DepTime: hora de sortida (hhmm)
- CRSDepTime: hora de sortida prevista (hhmm)
- ArrTime: hora d'arribada (hhmm)
- CRSArrTime: hora d'arribada prevista (hhmm)
- UniqueCarrier: codi del transportista
- FlightNum: número del vol
- TailNum: número de cua de l'avió
- ActualElapsedTime: temps transcorregut en minuts. Resultat de sumar les variables: AirTime, TaxiIn i TaxiOut.
- CRSElapsedTime: temps transcorregut previst en minuts.
- AirTime: duració del vol en minuts.
- ArrDelay: Retard respecte l'hora prevista d'arribada, en minuts. 
- DepDelay: Retard respecte l'hora de sortida prevista, en minuts.
- Origin: Codi de l'aeroport d'origen.
- Dest: Codi de l'aeroport de destí.
- Distance: Distancia entre el punt d'origen i de destí, en milles.
- TaxiIn: en minuts.
- TaxiOut: en minuts.
- Cancelled: si l'avió es va cancel·lar. 0 quan no es va cancel·lar i 1 en cas que es cancelés.
- CancellationCode: raó de la cancel·lació del vol. A = carrier, B = weather, C = NAS i D = security.
- Diverted: si l'avió es va desviar. 1 en cas que si, i 0 en cas que no. 
- CarrierDelay: en minuts. Retard associat al transportista.
- WeatherDelay: en minuts. Retard associat a condicions meteorològiques.
- NASDelay: en minuts. Retard associat al Sistema Nacional de l'Espai Aèri. 
- SecurityDelay: en minuts. Retard associat a la seguretat.
- LateAircraftDelay: en minuts. Retard associat al retard del mateix avió a l'anterior destació. Aquest efecte es coneix com la propagació del retard.

#### 2.5 Quins són els vols més llargs?

```python
print("Informació dels vols més llargs:\n")

longest_flights = flights.sort_values(by='Distance', ascending=False)
longest_flights[['Distance', 'UniqueCarrier', 'FlightNum', 'TailNum', 'Origin', 'Dest']].head()
```

#### I els més endarrerits? 

```python
print("Informació dels vols més endarrerits:\n")

highest_delays = flights.sort_values(by='ArrDelay', ascending=False)
highest_delays[['ArrDelay', 'UniqueCarrier', 'FlightNum', 'TailNum', 'Origin', 'Dest']].head()
```

Aquí veiem els vols més endarrerits. El major retard és de 1490.0 minuts.

#### Busca les rutes més llargues

```python
longest_routes = longest_flights[['Origin', 'Dest', 'Distance']]

# Ordenem les columnes 'Origin' i 'Dest' per evitar que les rutes en direcció inversa es considerin com a dues rutes diferents.
longest_routes[['Origin', 'Dest']] = longest_routes.apply(lambda row: sorted([row['Origin'], row['Dest']]), axis=1, result_type='expand')

longest_routes = longest_routes.drop_duplicates()

print('Rutes més llargues:\n')
longest_routes.head()
```

Tal i com podem veure les dues rutes més llargues tenen com a punt d'origen o destí Honolulu.

#### Busca les rutes que acumulen més retards¶

```python
delays_per_route = highest_delays.groupby(['Origin', 'Dest'])['ArrDelay'].sum().reset_index()

highest_delays[['Origin', 'Dest']] = highest_delays.apply(lambda row: sorted([row['Origin'], row['Dest']]), axis=1, result_type='expand')

delays_per_route = delays_per_route.drop_duplicates().sort_values(by='ArrDelay', ascending=False)

print('Retards acumulats per ruta:\n')
delays_per_route.head()
```

La ruta de l'aeroport LaGuardia (Nova York) a l'aeroport Internacional O'Hare (Chicago), en els dos sentits, és la que acumula més retards.

#### 2.6 Aporta allò que consideris rellevant.

```python
print('Número de retards segons el dia de la setmana:\n')

delays_per_dow = flights.groupby('DayOfWeek')['Delay'].sum().to_frame().rename(columns={'Delay': 'Number of Delays'}).sort_values(by='Number of Delays', ascending=False)
delays_per_dow
```

El dia amb més retards és el divendres, seguit del dilluns. En canvi, el dia amb menys retards és el dissabte.

Utilitzem la suma dels retards ja que quan hi ha hagut un retard 'Delay = 1', mentre que quan es considera que l'avió ha arribat en hora 'Delay = 0'. Per tant, amb la suma d'aquesta variable que hem creat podem obtenir el número de vols que han arribat amb retard. 

```python
print('Contribució de cada causa al retard:\n')

flights_with_delays = flights[flights['Delay'] == 1]
delay_data = flights_with_delays[['ArrDelay','CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']]
delay_proportions = delay_data.div(delay_data['ArrDelay'], axis=0) * 100
delay_proportions.mean().to_frame().rename(columns={0: 'Delay_contribution'})
```

Les causes que més contribueixen al retard en els vols són: la propagació del retard (degut al retard en el vol anterior), amb un 40%, i el retard associat a l'aerolínia, amb un 32,5%. La seguretat és, amb diferència, la causa que menys contribueix al retard total, amb menys d'un 1%.

```python
print('Vols en hora i retards per aerolínia:\n')

delays_per_carrier_data = {'UniqueCarrier': ['MQ', 'WN', 'DL', 'XE', 'EV', '9E', 'OO', 'FL', 'UA', 'AS', 'YV',
       'B6', 'AA', 'NW', 'HA', 'US', 'OH', 'F9', 'CO', 'AQ'],
                      'Delay': [0, 1]}

delays_per_carrier = flights.pivot_table(index='UniqueCarrier', columns='Delay', aggfunc='size', fill_value=0)
delays_per_carrier.columns = ['On Time', 'Delay']
delays_per_carrier.sort_values(by='Delay', ascending=False)
```
